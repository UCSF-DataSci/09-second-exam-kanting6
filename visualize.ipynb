# Walking speed analysis:

### Scatter plot of age vs walking speed with regression line
### Box plots by education level
### Line plot showing education age interaction

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load your CSV file
data = pd.read_csv('/Users/kankantingting/09-second-exam-kanting6/ms_data.csv')

# Scatter plot of age vs walking_speed with Regression Line
plt.figure(figsize=(10, 6))
sns.regplot(x='age', y='walking_speed', data=data, scatter_kws={'s': 20}, line_kws={'color': 'red'})
plt.title('Scatter plot of age vs walking_speed')
plt.xlabel('age')
plt.ylabel('walking_speed')
plt.show()

# Box plot of walking_speed by education_level
plt.figure(figsize=(10, 6))
sns.boxplot(x='education_level', y='walking_speed', data=data)
plt.title('Box plot of walking_speed by education_level')
plt.xlabel('education_level')
plt.ylabel('walking_speed')
plt.show()

# Line plot showing interaction of education_level and age on walking_speed
plt.figure(figsize=(12, 8))
sns.lineplot(x='age', y='walking_speed', hue='education_level', data=data, marker='o')
plt.title('Interaction of education_level and age on walking_speed')
plt.xlabel('age')
plt.ylabel('walking_speed')
plt.legend(title='education_level')
plt.show()
```

# Cost analysis:

### Bar plot of mean costs by insurance type
### Box plots showing cost distributions
### Add error bars or confidence intervals

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the insurance costs
insurance_costs = {
    "Basic": 100,
    "Premium": 150,
    "Platinum": 200
}

data = pd.DataFrame({
    'insurance_type': ['Basic', 'Premium', 'Platinum', 'Basic', 'Premium', 'Platinum'],
    'cost': [95, 140, 210, 105, 155, 195]
})

# Bar plot of mean costs by insurance type with error bars
plt.figure(figsize=(10, 6))
sns.barplot(x='insurance_type', y='cost', data=data, errorbar="sd", capsize=0.2)
plt.title('Mean Costs by Insurance Type with Error Bars')
plt.xlabel('Insurance Type')
plt.ylabel('Mean Cost')
plt.show()

# Box plot showing cost distributions by insurance type
plt.figure(figsize=(10, 6))
sns.boxplot(x='insurance_type', y='cost', data=data)
plt.title('Cost Distributions by Insurance Type')
plt.xlabel('Insurance Type')
plt.ylabel('Cost')
plt.show()
```

# Combined visualizations:

### Pair plot of key variables
### Faceted plots by education/insurance
### Time trends where relevant

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'age': [25, 30, 35, 40, 45, 50],
    'walking_speed': [1.2, 1.4, 1.3, 1.5, 1.4, 1.6],
    'education': ['High School', 'College', 'College', 'Graduate', 'Graduate', 'High School'],
    'insurance_type': ['Basic', 'Premium', 'Platinum', 'Basic', 'Premium', 'Platinum'],
    'date': pd.date_range(start='2023-01-01', periods=6, freq='ME')
})

### Pair Plot of Key Variables
sns.pairplot(data, vars=['age', 'walking_speed'], hue='insurance_type')
plt.suptitle('Pair Plot of Age and Walking Speed by Insurance Type', y=1.02)
plt.show()

### Faceted Plots by Education and Insurance Type
# Facet by insurance type for age vs. walking speed
g = sns.FacetGrid(data, col='insurance_type', height=4, aspect=1)
g.map_dataframe(sns.scatterplot, x='age', y='walking_speed')
g.set_axis_labels('Age', 'Walking Speed')
g.add_legend()
plt.suptitle('Age vs Walking Speed Faceted by Insurance Type', y=1.05)
plt.show()

# Facet by education for age vs. walking speed
g = sns.FacetGrid(data, col='education', height=4, aspect=1)
g.map_dataframe(sns.scatterplot, x='age', y='walking_speed')
g.set_axis_labels('Age', 'Walking Speed')
g.add_legend()
plt.suptitle('Age vs Walking Speed Faceted by Education Level', y=1.05)
plt.show()

### Time Trends (if a date or time variable is available)
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='walking_speed', hue='insurance_type', data=data, marker='o')
plt.title('Time Trend of Walking Speed by Insurance Type')
plt.xlabel('Date')
plt.ylabel('Walking Speed')
plt.legend(title='Insurance Type')
plt.show()
```
